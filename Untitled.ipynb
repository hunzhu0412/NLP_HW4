```python
import numpy as np
import pandas as pd
import json
```

```python
def loadData(papers,path):
    file = open(path, 'r', encoding='utf-8')

    for line in file.readlines():
        dic = json.loads(line)
        papers.append(dic)
```

```python
papers = []

folders = ['AA','AB','AC','AD','AE','AF','AG','AH','AI','AJ','AK','AL','AM']
nums = ['00','01','02','03','04','05','06','07','08','09']
for i in range(10,100):
    nums.append(i.__str__())
srt = '../wiki_zh/'

for folder in folders:
    tmp = srt+folder
    for n in nums:
        if folder == 'AM' and n == '74':
            break
        path = tmp+'/wiki_'+n
        print(path)
        loadData(papers,path)
```

```python
wiki = pd.DataFrame(papers)
wiki.drop(['id','url'],inplace=True,axis=1)
wiki.head()
```

```python
from opencc import OpenCC

cc = OpenCC('s2t')
i = 0
for text in wiki['text']:
    wiki['text'][i] = cc.convert(text)
    i = i+1
wiki.head()
```

```python
wiki.drop(['title'],inplace=True,axis=1)
wiki.head()
```

```python
import jieba
jieba.set_dictionary('../dict.txt.big')
stopwords = [line.strip() for line in open("../stop_words.txt",encoding="utf-8").readlines()]
jieba.load_userdict("../new.txt")
```

```python
with open('../wiki_seg.txt', 'w', encoding='utf_8_sig') as output:
    with open('../zh_wiki.txt','r', encoding='utf_8_sig') as content :
        for line in content:
            words = jieba.cut(line, cut_all=False)
            for word in words:
                if word not in stopwords:
                    if word == " " or word == "\n": 
                        continue
                    else:
                        output.write(word +' ')
```

```python
wiki.to_csv('../wiki.csv', header=None, index=None, encoding='utf_8_sig')
```

```python
from gensim.models import word2vec


# Settings
seed = 666
sg = 0
window_size = 10
vector_size = 100
min_count = 1
workers = 8
epochs = 5
batch_words = 10000

train_data = word2vec.LineSentence('../wiki_seg.txt')
model = word2vec.Word2Vec(
    train_data,
    min_count=min_count,
    vector_size=vector_size,
    workers=workers,
    epochs=epochs,
    window=window_size,
    sg=sg,
    seed=seed,
    batch_words=batch_words
)

model.save('../word2vec.model')
```
